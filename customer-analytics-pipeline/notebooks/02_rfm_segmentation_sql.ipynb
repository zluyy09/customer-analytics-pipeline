# 02 — RFM Segmentation (SQL‑first)

```python

import duckdb, os, pandas as pd
BASE = os.path.abspath("..")
con = duckdb.connect(os.path.join(BASE, "data/processed/warehouse.duckdb"))
con.execute("CREATE SCHEMA IF NOT EXISTS analytics; SET schema='analytics';")
for name in ["customers","orders","order_items","products","events","campaigns","campaign_members"]:
    con.execute(f"CREATE OR REPLACE TABLE {name} AS SELECT * FROM read_csv_auto('../data/raw/{name}.csv', SAMPLE_SIZE=-1);")
rfm_sql = open("../sql/rfm.sql").read()
con.execute("CREATE OR REPLACE VIEW rfm_view AS " + rfm_sql)
seg_sql = open("../sql/segmentation.sql").read()
seg = con.execute(seg_sql).fetchdf()
seg.segment.value_counts()
```
